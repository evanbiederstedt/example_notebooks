```r
#
# Pi simulations! 
#
# Was lazy, swiped from here: http://probaperception.blogspot.com/2012/10/estimation-of-number-pi-monte-carlo.html
#
# surfaceCircle = (surfaceCircle / surfaceSquare) * surfaceSquare
# Estimate (surfaceCircle/surfaceSquare): uniformly distribute k random variables in the square and count how many of them are in the circle, 
# then we have a proxy of the quantity (surfaceCircle/surfaceSquare)
#
```

```r
simulation <- function(long){
  c <- rep(0, long)
  numberIn <- 0
  for(i in 1:long){
    x <- runif(2,-1,1)
    if(sqrt(x[1]*x[1] + x[2]*x[2]) <= 1){
      numberIn = numberIn + 1
    }
    prop <- numberIn / i
    piHat <- prop *4
    c[i] <- piHat
  }
  return(c)
}

size = 1000
res = simulation(size)
ini = 1
plot(res[ini:size], type = 'l')
lines(rep(pi, size)[ini:size], col = 'red')
```
